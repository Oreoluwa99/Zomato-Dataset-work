```python
# We use autopct to display the percent value using Python string formatting. 
# For example, autopct='%1.1f%%' means that for each pie wedge, the format string is '1.1f%'.
# The autopct also has 1.1 meaning the data value on the chat would be in one decimal place.
# To change the data value to 2 decimal place, you can use 1.2
# We could use the Using shadow=True which does not allow for white backgrounds
plt.pie(Country_counts, labels=Country_names, autopct='%1.2f%%')
```

```python
#We can group the data according to the sizes of Aggregate rating, Rating color, and Rating text
df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
#To give the dataframe indexes, we can use the reset_index() method
df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
#To remame the last column, use the .rename(columns={0: 'xxxx'}) method
#The reset is used to reset the index starting number
df_final_rating=df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0: 'Rating count'})
```
